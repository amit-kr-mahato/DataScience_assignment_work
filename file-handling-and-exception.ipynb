### 5table in 4 line space using loop and fuction

```python
def table(filename, number, lines):
    with open(filename, "a") as file:
        for i in range(1, lines + 1):
            spaces = " " * (i - 1) * 4  
            file.write(f"{spaces}{number} x {i} = {number * i}\n")



with open("data.txt", "r") as file:
    content = file.read()
    print(content)

table("data.txt", 5, 10)
```
