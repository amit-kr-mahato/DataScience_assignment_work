## operator


## Arithmetic

```python
a = 5
b = 2
print(a + b)   
print(a - b)   
print(a * b)   
print(a / b)   
print(a // b) 
print(a % b)   
print(a ** b) 
```

## Comparison

```python
x = 5
y = 3
print(x == y)  
print(x != y)  
print(x > y)   
print(x < y)   
print(x >= 5)  
print(y <= 5)  
```

## Logical

```python
x = 5
print(x > 3 and x < 10) 
print(x > 6 or x == 5)   
print(not(x > 3))        
```

## Assignment

```python
x = 5
x += 3  
x -= 2   
x *= 2   
print(x)
```

## Membership

```python
fruits = ["apple", "banana", "mango"]
print("apple" in fruits)     
print("grape" not in fruits)  
```

## Identity

```python
a = [1, 2, 3]
b = a
c = [1, 2, 3]

print(a is b)      
print(a is c)      
print(a is not c)  
```
