### 1. Stack
Using library

```python
stack = []
stack.append(10)
stack.append(20)
print(stack.pop())
```

 ## Without library


```python
class Stack:
    def __init__(self):
        self.data = []

    def push(self, x):
        self.data.append(x)

    def pop(self):
        return self.data.pop() if self.data else None

s = Stack()
s.push(10)
s.push(20)
print(s.pop())
```

### 2. Queue
Using library

```python
from collections import deque

queue = deque()
queue.append(10)
queue.append(20)
print(queue.popleft())
```

without libraries

```python
class Queue:
    def __init__(self):
        self.data = []

    def enqueue(self, x):
        self.data.append(x)

    def dequeue(self):
        return self.data.pop(0) if self.data else None

q = Queue()
q.enqueue(10)
q.enqueue(20)
print(q.dequeue())
```

### 3. Linked List
Using library

```python
from array import array

linked_list = array('i', [10, 20, 30])
print(linked_list)
```

 ### without libraries

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new = Node(data)
        if not self.head:
            self.head = new
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new

ll = LinkedList()
ll.append(10)
ll.append(20)
```

### TREE Without library

```python
class TreeNode:
    def __init__(self, data):
        self.data = data
        self.children = []

    def add_child(self, node):
        self.children.append(node)

root = TreeNode(1)
child1 = TreeNode(2)
child2 = TreeNode(3)
root.add_child(child1)
root.add_child(child2)
```
