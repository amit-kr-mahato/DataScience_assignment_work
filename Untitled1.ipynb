### LIST OPERATIONS

```python
my_list=[1,2,3,4]
my_list.append(5)
print(my_list)
```

```python
my_list=[1,2,3,4]
my_list.insert(2,10)
print(my_list)
```

```python
my_list=[1,2,3,4]
my_list.remove(3)
print(my_list)
```

```python
my_list=[1,2,3,4]
my_list.pop()
print(my_list)
```

```python
my_list=[1,2,3,4]
my_list.sort()
print(my_list)
```

```python
my_list=[1,2,3,4]
my_list.reverse()
print(my_list)
```

```python
my_list=[50,60,70,80,90,100]
print(my_list[:2:-2])
```

```python
import time

start = time.time()
for _ in range(5_00_000):
    a = [1, 2, 3, 4]
end = time.time()
print("list create:", end - start)

start = time.time()
for _ in range(5_00_000):
    a = (1, 2, 3, 4)
end = time.time()
print("tuple create:", end - start)
```

```python
import sys
a=[1,2,3,4,5]
b=(1,2,3,4,5)
print(sys.getsizeof(a))
print(sys.getsizeof(b))
```

```python
a={1,2,3}
b={3,4,5}

print(a.union(b))
print(a.intersection(b))
print(a.difference(b))
print(a.symmetric_difference(b))
```

### Array

```python
from array import array
arr=array('i',[1,2,3,4])
print(arr)
```

```python
from collections import Counter
c=Counter("banana")
print(c)
```

```python
from collections import defaultdict
d=defaultdict(int)
d["a"] +=1
print(d)
```

```python
from collections import namedtuple
point=namedtuple("point","x y")
p=point(10,20)
print(p)
```
