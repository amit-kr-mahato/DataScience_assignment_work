### 3. Dictionary of Students and Marks; Print Highest

```python
students = {
    "Ram": 78,
    "Sita": 85,
    "Hari": 92
}

top = max(students, key=students.get)
print(top, students[top])
```

### 4. Stack to Reverse a String

```python
s = input()
stack = []

for ch in s:
    stack.append(ch)

rev = ""
while stack:
    rev += stack.pop()

print(rev)
```

### 5. Remove Duplicates Using Set

```python
nums = list(map(int, input().split()))
unique = list(set(nums))
print(unique)
```

### 7. Priority Queue Using heapq

```python
import heapq

pq = []
heapq.heappush(pq, 30)
heapq.heappush(pq, 10)
heapq.heappush(pq, 20)

while pq:
    print(heapq.heappop(pq))
```
