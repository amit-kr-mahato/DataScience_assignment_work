### Stack (Scratch Implementation)

```python
class Stack:
    def __init__(self):
        self.stack = []
    def push(self, x):
        self.stack.append(x)
    def pop(self):
        return self.stack.pop() if self.stack else None
    def top(self):
        return self.stack[-1] if self.stack else None
    def show(self):
        return self.stack

s = Stack()
s.push(1)
s.push(2)
s.push(3)
print(s.show())
print(s.pop())
print(s.show())

```

### Queue (Scratch Implementation)

```python
class Queue:
    def __init__(self):
        self.queue = []
    def enqueue(self, x):
        self.queue.append(x)
    def dequeue(self):
        return self.queue.pop(0) if self.queue else None
    def front(self):
        return self.queue[0] if self.queue else None
    def show(self):
        return self.queue

q = Queue()
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
print(q.show())
print(q.dequeue())
print(q.show())


```

### 4. LINKED LIST

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
    def append(self, x):
        new_node = Node(x)
        if not self.head:
            self.head = new_node
        else:
            t = self.head
            while t.next:
                t = t.next
            t.next = new_node
    def show(self):
        res = []
        t = self.head
        while t:
            res.append(t.data)
            t = t.next
        return res

ll = LinkedList()
ll.append(1)
ll.append(2)
ll.append(3)
print(ll.show())

```

### 5. TREE (Binary Tree)

### 6. GRAPH

```python
from collections import deque

graph = {
    "A": ["B", "C"],
    "B": ["D", "E"],
    "C": ["F"],
    "D": [],
    "E": ["F"],
    "F": []
}

def bfs(start):
    visited = set()
    q = deque([start])
    while q:
        node = q.popleft()
        if node not in visited:
            visited.add(node)
            print(node, end=" ")
            q.extend(graph[node])

bfs("A")

```

### Graph (Scratch Class Implementation)

```python
class Graph:
    def __init__(self):
        self.graph = {}
    def add_edge(self, u, v):
        self.graph.setdefault(u, []).append(v)
        self.graph.setdefault(v, []).append(u)
    def show(self):
        return self.graph

g = Graph()
g.add_edge("A", "B")
g.add_edge("A", "C")
print(g.show())

```
