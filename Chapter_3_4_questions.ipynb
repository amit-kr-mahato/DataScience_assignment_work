### 1. Largest of Three Numbers Using a Function

```python
def largest(a, b, c):
    return max(a, b, c)

x = int(input())
y = int(input())
z = int(input())

print(largest(x, y, z))
```

### 2. Multiplication Table Highlighting Multiples of 5

```python
##### n = int(input())

for i in range(1, 11):
    result = n * i
    if result % 5 == 0:
        print(f"*{result}*")
    else:
        print(result)
```

### 3. Roll a Dice 10 Times and Count Number of 6

```python
import random

count = 0

for i in range(10):
    roll = random.randint(1, 6)
    if roll == 6:
        count += 1

print(count)
```

### 4. Count Vowels and Consonants Using a Function

```python
def count_letters(word):
    vowels = 0
    consonants = 0
    for ch in word.lower():
        if ch.isalpha():
            if ch in "aeiou":
                vowels += 1
            else:
                consonants += 1
    return vowels, consonants

word = input()
v, c = count_letters(word)
print(v)
print(c)
```

### 5. Simple Grade Calculator for Multiple Students

```python
def grade(mark):
    if mark >= 90:
        return "A"
    elif mark >= 75:
        return "B"
    elif mark >= 60:
        return "C"
    elif mark >= 40:
        return "D"
    else:
        return "F"

marks = list(map(int, input().split()))

for m in marks:
    print(grade(m))
```
